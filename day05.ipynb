```python
with open('day05-input01.txt') as f:
    data = f.read()

# data = """47|53
# 97|13
# 97|61
# 97|47
# 75|29
# 61|13
# 75|53
# 29|13
# 97|29
# 53|29
# 61|53
# 97|53
# 61|29
# 47|13
# 75|47
# 97|75
# 47|61
# 75|61
# 47|29
# 75|13
# 53|13

# 75,47,61,53,29
# 97,61,53,29,13
# 75,29,13
# 75,97,47,61,53
# 61,13,29
# 97,13,75,29,47
# """

rule_strs, update_strs = data.split('\n\n')

rules = set()
for rule_str in rule_strs.splitlines():
    p1, p2 = rule_str.split('|')
    rules.add((p1, p2))

counter = 0
for update_str in update_strs.splitlines():
    pages = update_str.split(',')

    is_correct = True
    for i in range(len(pages)):
        for j in range(i+1, len(pages)):
            p1 = pages[i]
            p2 = pages[j]

            if (p1, p2) not in rules:
                is_correct = False
                break
        
        if is_correct is False:
            break
    
    # we no longer care about correct entries
    if is_correct is True:
        continue

    for i in range(len(pages)):
        rule_counter = 0
        for j in range(len(pages)):
            if i == j:
                continue

            p1 = pages[i]
            p2 = pages[j]

            if (p1, p2) in rules:
                rule_counter += 1
        
        if rule_counter == len(pages)//2:
            counter += int(pages[i])


    if is_correct:
        counter += int(pages[len(pages)//2])

counter
```
